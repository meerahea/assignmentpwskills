Que1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

Ans.

```python
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset using Seaborn
titanic  = sns.load_dataset("titanic")

# Create a scatter plot using Plotly Express
scatter_plot = px.scatter(titanic, x="age", y="fare", title="Scatter Plot of Age vs Fare")
scatter_plot.show()
```

Que3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
Ans.
```

```python
import plotly.express as px
import pandas as pd

# Load the tips dataset
tips = px.data.tips()
```

```python
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Histogram of Total Bill by Gender",
    labels={"sex": "Sex", "total_bill": "Total Bill"},
)

# Show the plot
fig.show()
```

Que4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

Ans.

```python
import plotly.express as px
import pandas as pd

# Load the iris dataset
iris = px.data.iris()
```

```python
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset",
)

# Show the plot
fig.show()
```

Que5. What is Distplot? Using Plotly express, plot a distplot.

Ans.

For making histograms and density plots in Plotly Express, the distplot function has been changed with the histogram function. You can customize the histograms you make with the histogram function, including the option to overlay density curves.

Here's how to use Plotly's histogram function to make a density plot:

```python
import plotly.express as px

# Load an example dataset
tips = px.data.tips()

# Create a density plot using the histogram function
fig = px.histogram(tips,x="total_bill",marginal="rug",# Display individual data points on the margins
    title="Density Plot of Total Bill",
    labels={"total_bill": "Total Bill"},
    opacity=0.7,# Adjust opacity of the bars
)

# Show the plot
fig.show()
```
