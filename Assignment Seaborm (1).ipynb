Built on top of Matplotlib, Seaborn is a potent Python data visualization package. It offers a sophisticated user interface for producing interesting and useful statistical visuals. Following are five examples of plots you can make with Seaborn and their typical applications:
scatter plot:
Use: A scatter plot represents the combination of two variables by showing points on a two-dimensional plane. It helps to illustrate the connection between two continuous variables.
The Bar Plot:
Rectangular bars in bar plots show categorical data. They are frequently employed to illustrate the distribution of a categorical variable or to compare the values of various categories.
Line Plot:
Use: Line plots are used to show the development or trend of data points over a continuous period of time. They are especially beneficial for time series data, which depict changes in values over the course of time.
Histogram:
Use: Line plots are used to show the development or trend of data points over a continuous period of time. They are especially beneficial for time series data, which depict changes in values over the course of time.
Box Plot (Box and Whisker Plot):
Use: Box plots display the distribution of data through their quartiles. They show the median, upper and lower quartiles, and possible outliers. Box plots are great for identifying the spread and skewness of data, especially in comparison between different categories.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

Ans:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.legend(loc="upper right")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

Ans:

```python
import matplotlib.pyplot as plt
import seaborn as sns
#load data set of titanic
titanic = sns.load_dataset('titanic')
# a figure with subplots for boxes
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Box plot for 'age' vs 'pclass'
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
axes[0].set_title('Age vs Pclass')

# Box plot for 'fare' vs 'pclass'
sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])
axes[1].set_title('Fare vs Pclass')

# Adjust layout
plt.tight_layout()
plt.show()

```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Ans:

```python
import matplotlib.pyplot as plt
import seaborn as sns
#load data set of diamonds
diamonds= sns.load_dataset('diamonds')

# Create a histogram with hue for 'cut'
sns.histplot(data=diamonds, x='price', hue='cut',multiple='stack', kde=True)

# Set plot title and labels
plt.title('Histogram for Price by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

Ans:

```python
import matplotlib.pyplot as plt
import seaborn as sns

#load data set iris
iris= sns.load_dataset('iris')

#a pair plot with hue for 'species'
sns.pairplot(iris, hue='species')

#plot title
plt.title('Pair Plot of Iris Dataset by Species')

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Ans:

```python
import matplotlib.pyplot as plt
import seaborn as sns

#load data from flights
flights = sns.load_dataset('flights')

# Pivot the data to create a matrix suitable for a heatmap
flights_matrix = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_matrix , cmap='YlGnBu', annot=True, fmt='d', linewidths=0.6)

# Set plot title
plt.title('Flights Data Heatmap')

# Display the plot
plt.show()
```
